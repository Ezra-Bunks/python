```python
"""
Course: BSE2302	PROFESSIONAL SOFTWARE ENGINEERING MINI PRACTICAL PROJECT II
Assignment6: Customer Churn
Name: Mukisa Ben Ezra
RegNo: 24/U/21328
Date: June 30, 2026
"""

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 1. Create Sample Data
data = {
    'MonthlyCharge': [50, 80, 70, 40, 90, 60, 100, 85, 45, 75],
    'ContractLength': [12, 6, 24, 12, 3, 18, 24, 6, 12, 18],
    'Age': [30, 25, 40, 35, 28, 50, 45, 32, 29, 55],
    'Churn': [0, 1, 0, 0, 1, 0, 0, 1, 0, 0] # 0 = No Churn, 1 = Churn
}

# Create a DataFrame
df = pd.DataFrame(data)

print("--- Sample Data ---")
print(df)
print("\n")

# 2. Separate Features (X) and Target (y)
X = df[['MonthlyCharge', 'ContractLength', 'Age']] # Features
y = df['Churn'] # Target variable

# 3. Split Data into Training and Testing Sets (for demonstration)
# With very small datasets, splitting might not be ideal for real-world performance,
# but it's good practice for model evaluation.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("--- Training the Logistic Regression Model ---")
# 4. Initialize and Train the Logistic Regression Model
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
print("Model training complete.\n")

# 5. Make Predictions on the Test Set
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1] # Probability of churn

print("--- Model Predictions ---")
print(f"Actual Churn (Test Set):   {list(y_test)}")
print(f"Predicted Churn:           {list(y_pred)}")
print(f"Predicted Churn Probability: {[f'{p:.2f}' for p in y_prob]}\n")

# 6. Evaluate the Model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, zero_division=0)

print("--- Model Evaluation ---")
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", report)

# Example: Predict for a new customer
print("--- Predicting for a New Customer ---")
new_customer_data = pd.DataFrame([[70, 12, 35]], columns=['MonthlyCharge', 'ContractLength', 'Age'])
new_customer_prediction = model.predict(new_customer_data)
new_customer_probability = model.predict_proba(new_customer_data)[:, 1]

print(f"New Customer Data:\n{new_customer_data}")
print(f"Predicted Churn: {new_customer_prediction[0]} (0=No Churn, 1=Churn)")
print(f"Probability of Churn: {new_customer_probability[0]:.2f}")
```
